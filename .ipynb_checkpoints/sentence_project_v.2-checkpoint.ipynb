```python
"""
SENTENCE EXTRACTOR

WORD
Use def extract_sentence_word(word_list)

Use all word forms you need, e.g.
NOUNS: ["cat", cats"]
VERBS: ["drive", "drives", "drove", "driven", "driving"]
ADJECTIVES["big", "bigger", "biggest"] etc.
All INSEPARABLE phrasal verbs ["wake up"] and phrases ["make a decision"] go here

PHRASE
def extract_sentence_phrase(phrase_list)

Use all word forms you need
All SEPARABLE phrasal verbs ["turn [smth] on"] and phrases ["make up [one's] mind"] go here
"""

import re

with open("harry_potter_corpus.txt", "r") as handle:
    contents_of_book = handle.read()

contents_list = contents_of_book.split(".")

def extract_sentence_word(word_list):
    for word in word_list:
        print(word.upper())
        print()
        for sentence in contents_list:
            if " " + word + " " in sentence.lower() or sentence.lower().endswith(" " + word) or sentence.lower().startswith(word + " "):
                capitalised_word = word.upper()
                capitalised_sentence = sentence.replace(word, capitalised_word)
                gap_sentence = sentence.replace(word, "_____________")
                print(capitalised_sentence + ".")
                print(gap_sentence + ".")
                print()

def extract_sentence_phrase(phrase_list):
    for phrase in phrase_list:
        print(phrase[0].upper() + "..." + phrase[1].upper())
        print()
        for sentence in contents_list:
            regex_match = re.search("(\s|^)" + phrase[0] + "\s.{2,15}\s" + phrase[1] + "(\s|$)", sentence.lower())
            if regex_match is not None:
                if sentence.lower().startswith(phrase[0] + " "):
                    phrase_start = 0
                    phrase_end = regex_match.end()
                elif sentence.lower().endswith(" " + phrase[1]):
                    phrase_start = regex_match.start() + 1
                    phrase_end = regex_match.end() 
                else:
                    phrase_start = regex_match.start() + 1
                    phrase_end = regex_match.end() - 1
                capitalised_sentence = sentence[:phrase_start] + \
                                        phrase[0].upper() + \
                                        sentence[phrase_start + len(phrase[0]) : phrase_end - len(phrase[1])] + \
                                        phrase[1].upper() + \
                                        sentence[phrase_end:]
                gap_sentence = sentence[:phrase_start] + \
                                "_____________" + \
                                sentence[phrase_start + len(phrase[0]): phrase_end - len(phrase[1])] + \
                                "_____________" + \
                                sentence[phrase_end:]
                print(capitalised_sentence + ".")
                print(gap_sentence + ".")
                print()
                

extract_sentence_word(["make a decision", "made a decision", "making a decision"])
extract_sentence_phrase([("make up", "mind"), ("makes up", "mind"), ("made up", "mind")])
```
